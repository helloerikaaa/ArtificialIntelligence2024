```python
from collections import deque
```

```python
# Definición de la función para BFS
def bfs(graph, start):
    visited = set()            # Conjunto para almacenar nodos visitados
    queue = deque([start])     # Cola para almacenar nodos a visitar
    
    while queue:               # Mientras haya nodos en la cola
        node = queue.popleft() # Sacar el primer nodo de la cola
        if node not in visited:
            print(node)       # Procesar el nodo (en este caso, simplemente imprimirlo)
            visited.add(node) # Marcar el nodo como visitado
            queue.extend(graph[node] - visited) # Agregar los vecinos no visitados a la cola
```

```python
# Definición de la función para DFS
def dfs(graph, start):
    visited = set()            # Conjunto para almacenar nodos visitados
    stack = [start]            # Pila para almacenar nodos a visitar
    
    while stack:               # Mientras haya nodos en la pila
        node = stack.pop()     # Sacar el último nodo de la pila
        if node not in visited:
            print(node)       # Procesar el nodo (en este caso, simplemente imprimirlo)
            visited.add(node) # Marcar el nodo como visitado
            stack.extend(graph[node] - visited) # Agregar los vecinos no visitados a la pila
```

```python
# Ejemplo de un grafo representado como un diccionario de conjuntos de nodos vecinos
graph = {
    'A': {'B', 'C'},
    'B': {'A', 'D', 'E'},
    'C': {'A', 'F'},
    'D': {'B'},
    'E': {'B', 'F'},
    'F': {'C', 'E'}
}
```

```python
print("BFS:")
bfs(graph, 'A') # Llamada a la función bfs para iniciar desde el nodo 'A'
print("\nDFS:")
dfs(graph, 'A') # Llamada a la función dfs para iniciar desde el nodo 'A'
```
