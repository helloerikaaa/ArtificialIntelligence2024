```python
import nltk # Librería para NLP
from nltk.corpus import stopwords
from collections import Counter # Contar palabras
import string
```

```python
# Descargar los recursos necesarios para trabajar
# Con el español
nltk.download('punkt') # Paquete para tokenizar
nltk.download('stopwords')
```

```python
def contar_palabras(texto):
    # Pasar el texto a minúsculas
    texto_min = texto.lower()

    # Eliminar las puntuaciones del texto
    texto_limpio = texto_min.translate(
        str.maketrans('', '', string.punctuation))

    # Eliminar stopwords y palabras no importantes
    # Tokenizar en palabras
    palabras = nltk.word_tokenize(texto_limpio)

    # Obtener la lista de stopwords
    stop_words = set(stopwords.words('spanish'))

    # Filtrar las palabras
    palabras_filtradas = [palabra for palabra 
                          in palabras if palabra 
                          not in stop_words]
    # Contar palabras
    conteo_palabras = Counter(palabras_filtradas)

    # Top 10 de palabras más utilizadas
    top10 = conteo_palabras.most_common(10)

    print(f"Texto Original : {texto}")
    print(f"Texto Procesado: {palabras_filtradas}")
    print("Palabras más utilizadas:")
    for palabra, frecuencia in top10:
        print(f"{palabra}: {frecuencia}")
```

```python
frase = "Hoy es lunes y nadie tiene ganas de clase. Arriba el pollo rostizado. Hoy refugio no tiene empatía"
```

```python
contar_palabras(frase)
```
